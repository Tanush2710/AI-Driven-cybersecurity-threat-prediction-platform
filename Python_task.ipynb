Setup & Basic DataFrame

```python
import pandas as pd

# Create a simple dataset
data = {
    "user_id": [1,2,3,4],
    "name": ["Aisha","Rahul","Meera","Arjun"],
    "age": [28,22,31,24],
    "city": ["Gwalior","Bhopal","Indore","Gwalior"]
}

df = pd.DataFrame(data)
df
```

Basic Data Analysis Tasks

```python
# View summary statistics
print(df.describe())

# Select specific columns
print(df[['name', 'city']])

# Filter rows
print(df[df['age'] > 25])

# Sort
print(df.sort_values('age'))
```

Data Cleaning Tasks

```python
# Add missing value
df.loc[2, 'city'] = None

# Handle missing values
df['city'] = df['city'].fillna("Unknown")

# Change data types
df['age'] = df['age'].astype(int)

df
```

Simple Calculator (Functions)

```python
def add(a,b): return a+b
def sub(a,b): return a-b
def mul(a,b): return a*b
def div(a,b):
    if b == 0:
        return "Error: Divide by zero"
    return a/b

print(add(5,3))
print(div(10,0))
```

Calculator Class

```python
class Calculator:
    def add(self,a,b): return a+b
    def sub(self,a,b): return a-b
    def mul(self,a,b): return a*b
    def div(self,a,b):
        if b == 0: return "Error"
        return a/b

calc = Calculator()
print(calc.add(10, 5))
print(calc.mul(4, 3))
```

Simple Plot

```python
import matplotlib.pyplot as plt

plt.bar(df['name'], df['age'])
plt.xlabel("Users")
plt.ylabel("Age")
plt.title("Age of Users")
plt.show()
```

File Handling Task

```python
# Save as CSV
df.to_csv("users.csv", index=False)

# Read again
new_df = pd.read_csv("users.csv")
new_df
```

Join Two DataFrames

```python
# Another simple table
orders = pd.DataFrame({
    "order_id": [101,102,103],
    "user_id": [1,1,3],
    "amount": [500,1200,800]
})

# INNER JOIN
merged = df.merge(orders, on="user_id", how="inner")
merged
```

GroupBy Task

```python
# Total amount spent per user
print(merged.groupby("name")["amount"].sum())
```
